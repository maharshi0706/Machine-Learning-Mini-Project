```python
import pandas as pd
import numpy as np
import seaborn as sns
import glob
import os

df = pd.concat(
    map(pd.read_csv, glob.glob(os.path.join("../Datasets/","*.csv" ))), ignore_index= True)
df
```

```python
df.rename(columns={'FTHG':'FTHomeGoals','FTAG':'FTAwayGoals','FTR':'FTResult','HTR':'HTResult','HTHG':'HTHomeGoals','HTAG':'HTAwayGoals'
                        ,'HS':'HomeShots','AS':'AwayShots','HST':'HomeShotsTarget','HHW':'HomeWoodwork','AHW':'AwayWoodwork',
                        'HC':'HomeCorners','AC':'AwayCorners','HF':'HomeFouls','AF':'AwayFouls','HFKC':'AwayFreekicks'
                        ,'AFKC':'HomeFreekicks','HO':'HomeOffsides','AO':'AwayOffsides','HY':'HomeYellow','AY':'AwayYellow',
                        'HR':'HomeRed','AR':'AwayRed','HBP':'HomeBookingPoints','ABP':'AwayBookingPoints'},inplace=True)
```

```python
df.drop(df.iloc[:,22:127], inplace=True, axis=1)
# df.drop(['B365H'])
df
```

```python
numerical = df.select_dtypes(include=['int', 'float']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Impute missing values for numerical features with mean
for col in numerical:
    mean_value = df[col].mean()
    df[col].fillna(mean_value, inplace=True)

# Impute missing values for categorical features with mode
for col in categorical_cols:
    mode_value = df[col].mode().iloc[0]  # Get the most frequent category
    df[col].fillna(mode_value, inplace=True)

df
```

```python
import matplotlib.pyplot as plt
df.duplicated().any()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
# Impute missing values for categorical features with mode
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
```

```python
# Outlier detection and handling
from scipy.stats import skew

numerical = df.select_dtypes(include=['int64','float64'])

skewness = numerical.apply(lambda x: skew(x.dropna()))

zScore = np.abs((numerical - numerical.mean()) / numerical.std())
cleanDF = df[(zScore < 3).all(axis=1)]
print('Shape :',df.shape)
print('Shape :',cleanDF.shape)

zScore
```

```python
# Model Training
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
X = cleanDF.drop(['FTResult','Date','Referee','HomeTeam','AwayTeam','Div'], axis=1)
y = cleanDF['FTResult']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20)
X.head()
```

```python
# RANDOM FOREST
from sklearn.ensemble import RandomForestClassifier

randomForest = RandomForestClassifier(n_estimators=100, random_state=42)

randomForest.fit(X_train, y_train)

y_pred = randomForest.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print ('Accuracy: ' , accuracy*100,'%')
```

```python
# Support Vector Machine
from sklearn.svm import SVC
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

svm = SVC(kernel='linear', decision_function_shape='ovo')
svm.fit(X_train, y_train)

y_pred = svm.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print ('Accuracy: ' , accuracy*100,'%')
print("Classification Report:")
print(classification_report(y_test, y_pred))

k = 10
kf = KFold(n_splits=k, shuffle=True, random_state=30)

scores = cross_val_score(svm, X, y, cv=kf)

avg_score = sum(scores) / len(scores)
avg_score
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier

decisionTree = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=42)
decisionTree.fit(X_train, y_train)

y_pred = decisionTree.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print ('Accuracy: ' , accuracy*100,'%')
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score


logisticRegression = LogisticRegression(max_iter=30000,multi_class='ovr')
logisticRegression.fit(X_train, y_train)


y_pred = logisticRegression.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print ('Accuracy: ' , accuracy*100,'%')

k = 10
kf = KFold(n_splits=k, shuffle=True, random_state=30)

scores = cross_val_score(logisticRegression, X, y, cv=kf)

avg_score = sum(scores) / len(scores)
avg_score
```
